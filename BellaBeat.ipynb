```python
person_df = df.groupby('Id').agg({
    'TotalSteps': 'sum',
    'VeryActiveMinutes': 'sum',
    'FairlyActiveMinutes': 'sum',
    'LightlyActiveMinutes': 'sum',
    'SedentaryMinutes': 'sum',
    'Calories': 'sum'
       # total activity
}).reset_index()
```

```python
person_df['activity_group'] = pd.cut(
    person_df['TotalSteps'],
    bins=[0, 120000, 240000, 360000, float('inf')],
    labels=['Low', 'Moderate', 'High', 'Very High']
)
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/minuteSleep_merged.csv")
```

```python
df['ActivityDate'] = pd.to_datetime(df['ActivityDate']).dt.date
df2['date'] = pd.to_datetime(df2['date']).dt.date
merged_df = pd.merge(df, df2, left_on=['Id', 'ActivityDate'], right_on=['Id', 'date'])
```

```python
merged_df.shape
merged_df.head()
merged_df.isnull().sum()
```

```python
merged_df.describe()
```

```python
sleep_level = merged_df.groupby(['Id','logId','value']).size().unstack(fill_value=0)
```

```python
sleep_level.columns = ['asleep', 'restless', 'awake']
```

```python
sleep_level['total_sleep'] = sleep_level['asleep']
```

```python
sleep_level['efficiency'] = sleep_level['asleep'] / (
    sleep_level['asleep'] + sleep_level['restless'] + sleep_level['awake']
)
```

```python
sleep_level_data = sleep_level.reset_index()[['Id', 'logId', 'efficiency']]
merged_df = pd.merge(merged_df, sleep_level_data, on=['Id', 'logId'], how='left')
merged_df[['efficiency','TotalSteps']].corr()
```

```python
merged_df = pd.merge(merged_df, person_df[['Id', 'activity_group']], on='Id', how='left')
merged_df.groupby('activity_group')['efficiency'].mean()
```

```python


group_eff = merged_df.groupby('activity_group')['efficiency'].mean()

group_eff.plot(kind='bar')

plt.title('Sleep Efficiency by Activity Group')
plt.ylabel('Efficiency')
plt.show()
```
